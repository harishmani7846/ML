```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
IMG_SIZE = 224
BATCH_SIZE = 32
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)
```

```python
train_generator = train_datagen.flow_from_directory(
    '/content/drive/MyDrive/mri',
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='training'
)
```

```python
val_generator = train_datagen.flow_from_directory(
    '/content/drive/MyDrive/mri',
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='validation'
)
```

```python
model = keras.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train
model.fit(train_generator, epochs=4, validation_data=val_generator)

# Save the model
model.save('/content/drive/MyDrive/brarin/mri_model.h5')
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np
```

```python
model = load_model('/content/drive/MyDrive/brarin/mri_model.h5')
print("Model Loaded")
```

```python
test_img_path = "/content/drive/MyDrive/mri/test/yes/Y1.jpg"  # example image
img = image.load_img(test_img_path, target_size=(224, 224))
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0) / 255.0

prediction = model.predict(img_array)

```

```python
if prediction >= 0.5:
    print("Tumor Detected")
else:
    print("No Tumor")
```
