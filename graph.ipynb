```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips =sns.load_dataset("tips")
```

```python
sns.barplot(x="day",y="total_bill",data=tips)
plt.title("total bill day")
plt.xlabel("day")
plt.ylabel("total bil")
plt.show()
```

```python
sns.scatterplot(x="day",y="total_bill",data=tips)
plt.title("total bill day")
plt.xlabel("day")
plt.ylabel("total bil")
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=tips)
plt.title("total bill day")
plt.xlabel("day")
plt.ylabel("total bil")
plt.show()
```

```python
sns.barplot(x="day",y="total_bill",data=tips)
plt.title("total bill day")
plt.xlabel("day")
plt.ylabel("total bil")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=tips)
plt.title("total bill day")
plt.xlabel("day")
plt.ylabel("total bil")
plt.show()
```

```python
sns.histplot(x="day",y="total_bill",data=tips)
plt.title("total bill day")
plt.xlabel("day")
plt.ylabel("total bil")
plt.show()
```

```python
iris=sns.load_dataset("iris")
# Exclude the 'species' column before calculating the correlation
numerical_iris = iris.drop('species', axis=1)
correlation = numerical_iris.corr()
sns.heatmap(correlation,annot=True,cmap="coolwarm")
plt.title("iris")
plt.show()
```

```python
sns.pairplot(iris,hue="species")
plt.show()
```
