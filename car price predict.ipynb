```python

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
```

```python
car = pd.read_csv('/content/car_price_prediction.csv')
```

```python
car.head()
```

```python
car.isnull().sum()
```

```python
le_fuel = LabelEncoder()
le_transmission = LabelEncoder()
le_brand = LabelEncoder()
le_model = LabelEncoder()
le_condition = LabelEncoder()

car['Fuel_Type_n'] = le_fuel.fit_transform(car['Fuel Type'])
car['Transmission_n'] = le_transmission.fit_transform(car['Transmission'])
car['Brand_n'] = le_brand.fit_transform(car['Brand'])
car['Model_n'] = le_model.fit_transform(car['Model'])
car['Condition_n'] = le_condition.fit_transform(car['Condition'])
```

```python
a = car[['Car ID', 'Brand_n', 'Year', 'Fuel_Type_n', 'Transmission_n', 'Mileage', 'Condition_n', 'Model_n']]
b = car['Price']
```

```python
lr = LinearRegression()
```

```python
lr.fit(a, b)
```

```python
r = lr.predict([[10, 5, 2020, 1, 0, 15000, 2, 12]])
```

```python
sample_input = pd.DataFrame([[10, 5, 2020, 1, 0, 15000, 2, 12]],
                            columns=['Car ID', 'Brand_n', 'Year', 'Fuel_Type_n',
                                     'Transmission_n', 'Mileage', 'Condition_n', 'Model_n'])
```

```python
r = lr.predict(sample_input)

```

```python
print("Predicted Price:", r[0])
```
